```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
import glob
```

```python
# manually set main directory and catalog files locations
PRIMARY_DIR  = "/Users/research/projects/alderaan/"

QUICK_TTV_DIR = PRIMARY_DIR + "QuickTTVs/"
SIM_TTV_DIR   = PRIMARY_DIR + "Simulations/TTVs/"
```

```python
truth_files = np.sort(glob.glob(SIM_TTV_DIR + "S*"))
map_files = np.sort(glob.glob(QUICK_TTV_DIR + "S*/*map*"))
smooth_files = np.sort(glob.glob(QUICK_TTV_DIR + "S*/*smooth*"))
quick_files = np.sort(glob.glob(QUICK_TTV_DIR + "S*/*quick*"))
```

```python
truth_planet_ids = []
map_planet_ids = []
smooth_planet_ids = []
quick_planet_ids = []

for i, tf in enumerate(truth_files):
    truth_planet_ids.append(tf[-22:-13])
    
for i, mf in enumerate(map_files):
    map_planet_ids.append(mf[-24:-15])

for i, sf in enumerate(smooth_files):
    smooth_planet_ids.append(sf[-24:-15])
    
for i, qf in enumerate(quick_files):
    quick_planet_ids.append(qf[-24:-15])


    
# only use the planets for which Quick TTVs have been computed
match = np.isin(truth_planet_ids, quick_planet_ids)

truth_files = truth_files[match]
truth_planet_ids = np.array(truth_planet_ids)[match]
```

```python
koi_ids = []

for i, pid in enumerate(truth_planet_ids):
    koi_ids.append("S" + pid[1:-3])
```

```python
for i, koi in enumerate(np.unique(koi_ids)):
    use = np.array(koi_ids) == koi
    NPL = np.sum(use)


    tfiles = truth_files[use]
    mfiles = map_files[use]
    sfiles = smooth_files[use]

    for npl in range(NPL):
        tinds, ttts = np.loadtxt(tfiles[npl]).swapaxes(0,1)
        minds, mtts = np.loadtxt(mfiles[npl]).swapaxes(0,1)
        sinds, stts = np.loadtxt(sfiles[npl]).swapaxes(0,1)
        
        tinds = np.asarray(tinds, dtype="int")
        minds = np.asarray(minds, dtype="int")
        sinds = np.asarray(sinds, dtype="int")

        t0, per = poly.polyfit(tinds, ttts, 1)

        minds += int(np.round((mtts.min()-ttts.min())/per)) 
        sinds += int(np.round((stts.min()-ttts.min())/per)) 

        tephem = poly.polyval(tinds, [t0,per])
        mephem = poly.polyval(minds, [t0,per])
        sephem = poly.polyval(sinds, [t0,per])

        tomc = ttts - tephem
        momc = mtts - mephem
        somc = stts - sephem

        fig, ax = plt.subplots(2,1, figsize=(12,6))
        
        ax[0].plot(tephem, tomc*24*60, "k", lw=3, label="Truth")
        ax[0].plot(mephem, momc*24*60, "o", c="orange", label="MAP")
        ax[0].legend(fontsize=16)

        ax[1].plot(tephem, tomc*24*60, "k", lw=3, label="Truth")
        ax[1].plot(sephem, somc*24*60, "o", c="cornflowerblue", label="Smooth")
        ax[1].legend(fontsize=16)
        
        
    plt.suptitle(koi, fontsize=24)
    plt.show()
    
    print("MAP RMS =", np.round(np.std(mtts-ttts[minds])*24*60, 1))
    print("Smooth RMS =", np.round(np.std(stts-ttts[sinds])*24*60, 1))
    print("Linear RMS =", np.round(np.std(tomc[sinds])*24*60, 1))
```
