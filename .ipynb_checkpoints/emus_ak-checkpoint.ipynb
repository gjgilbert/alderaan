```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
import scipy.stats as stats
from scipy.special import logsumexp
import scipy.linalg as linalg
from astropy.stats import freedman_bin_width

pi = np.pi
```

# First simulate a complicated potential

```python
Npdf = int(1e6)
random = np.random.seed(222)
#random = np.random.seed(234)

xpts = np.random.uniform(0,10, size=6)
ypts = np.random.uniform(3,6, size=6)

pfit = poly.polyfit(xpts,ypts,5)

x_pdf = np.linspace(0,10,Npdf)
y_pdf = poly.polyval(x_pdf, pfit)
#y_pdf = y_pdf - y_pdf.min() + 0.1
y_pdf = y_pdf/(np.sum(y_pdf)/Npdf*(x_pdf.max()-x_pdf.min()))

plt.plot(x_pdf, y_pdf);
```

# Define umbrella bias windows

```python
def psi_0(x):
    return 2*stats.triang(c=0.5, loc=-4, scale=8).pdf(x)

def psi_1(x):
    return stats.triang(c=0.5, loc=1, scale=8).pdf(x)

def psi_2(x):
    return 2*stats.triang(c=0.5, loc=6, scale=8).pdf(x)
```

```python
psi_fxns = [psi_0, psi_1, psi_2]

Nwin = len(psi_fxns)

for i in range(Nwin):
    plt.plot(x_pdf, psi_fxns[i](x_pdf))
```

# Draw (biased) samples from each window

```python
Nsamp = Npdf//10
samples = np.zeros((Nwin,Nsamp))

for i, psi in enumerate(psi_fxns):
    prob = np.copy(psi(x_pdf)*y_pdf)
    # prob = np.copy(y_pdf)
    prob /= np.sum(prob)
    
    samples[i] = np.random.choice(x_pdf, p=prob, size=Nsamp)
    

bins = np.linspace(0,10,41)

fig, ax = plt.subplots(1,2, figsize=(12,4))

for i, samp in enumerate(samples):
    ax[0].hist(samp, bins=bins, density=True, histtype="step", lw=2)
    
ax[1].hist(np.hstack(samples), bins=bins, density=True, histtype="step", lw=2, color="grey")
ax[1].plot(x_pdf, y_pdf, "r")
```

```python
from scipy.linalg import qr
from scipy.linalg import inv
from scipy.linalg import solve
import numpy as np


def stationary_distrib(F,residtol = 1.E-10,max_iter=100):
    """
    Calculates the eigenvector of the matrix F with eigenvalue 1 (if it exists).
    
    Parameters
    ----------
    F : ndarray
        A matrix known to have a single left eigenvector with 
        eigenvalue 1.
    
    residtol : float or scalar
        To improve the accuracy of the computation, the algorithm will
        "polish" the final result using several iterations of the power
        method, z^T F = z^T.  Residtol gives the tolerance for the 
        associated relative residual to determine convergence.
    maxiter : int
        Maximum number of iterations to use the power method to reduce
        the residual.  In practice, should never be reached.
    
    Returns
    -------
    z : ndarray
        The eigenvector of the matrix F with eigenvalue 1.  For a Markov
        chain stationary distribution, this is the stationary distribution.
        Normalization is chosen s.t. entries sum to one.
    
    """ 
    
    L = len(F) # Number of States
    M = np.eye(L)-F
    q,r=qr(M)
    z=q[:,-1] # Stationary dist. is last column of QR fact
    z/=np.sum(z) # Normalize Trajectory 
    # Polish solution using power method.
    for itr in range(max_iter):
        znew = np.dot(z,F)
        tv = np.abs(znew[ z>0 ] - z[ z>0 ]) 
        tv = tv/z[z > 0 ]
        maxresid = np.max(tv) # Convergence Criterion 
        if maxresid < residtol:
            break
        else:
            z = znew

    return z/np.sum(z) # Return normalized (by convention)
```

```python
def estimate_F(psi_fxns, samples, z):
    nwin = len(psi_fxns)
    F = np.zeros((nwin,nwin))
    for i in range(nwin):
        denom = 0.
        for k in range(nwin):
            denom += psi_fxns[k](samples[i])/z[k]
        for j in range(nwin):
            F[i,j] = np.mean(psi_fxns[j](samples[i])/z[i] / denom)
    
    return F
```

```python
z = np.zeros(Nwin)

for i, psi in enumerate(psi_fxns):
    z[i] = np.mean(psi(samples[i]))
    
z /= np.sum(z)
```

```python
# get an estimate of F
F = estimate_F(psi_fxns, samples, z)

# solve for z using usample function 
z = stationary_distrib(F)
```

```python
combo_samples = []
weights = []
for i in range(Nwin):
    combo_samples.append(samples[i])
    
    # compute weights a in eq 4 in Matthews et al. 
    wd = 0.
    for k in range(Nwin):
        wd += psi_fxns[k](samples[i])/z[k]

    weights.append(1./wd)
combo_samples = np.array(combo_samples).flatten()
weights = np.array(weights).flatten()

plt.hist(combo_samples, weights=weights, bins=bins, histtype="step", density=True, lw=2, color="grey", label='samples histogram')
plt.plot(x_pdf, y_pdf, label='target odf')
plt.legend(frameon=False, loc='best')
plt.show()
```
