```python
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
import pymc3_ext as pmx
import arviz as az
import exoplanet as xo
import corner
import aesara_theano_fallback.tensor as T
import scipy.signal as sig

pi = np.pi
```

```python
N = 10000

r_min = 1e-3
r_max = 1.0
```

# Draw from xo.ImpactParameter

```python
with pm.Model():

    r = pm.Uniform("r", lower=r_min, upper=r_max)
    b = xo.ImpactParameter("b", ror=r)

    trace_rand = pm.sample(draws=N, return_inferencedata=True)
```

```python
corner.corner(trace_rand);
```

# Simulate a transit and sample w/ DFM's parameterization

```python
t = np.sort(np.random.default_rng(678).uniform(-0.1, 0.1, 500))
yerr = 10.0

r_min, r_max = 0.01, 0.5

with pm.Model() as model_dfm:

    mean = pm.Normal("mean")
    log_jit = pm.Normal("log_jit", mu=0.0, sigma=5.0)
    
    u = xo.QuadLimbDark("u")
    #log_r = pm.Uniform("log_r", lower=np.log(r_min), upper=np.log(r_max), testval=np.log(0.1))
    #r = pm.math.exp(log_r)
    r = pm.Uniform("r", lower=r_min, upper=r_max, testval=0.1)
    b = xo.ImpactParameter("b", ror=r, testval=0.86)
    
    t0 = pm.Normal("t0", sigma=0.1)
    log_duration = pm.Uniform("log_duration", lower=np.log(0.01), upper=np.log(0.2), testval=np.log(0.1))
    
    starrystar = xo.LimbDarkLightCurve(u)
    orbit = xo.orbits.KeplerianOrbit(period=10.0, t0=t0, b=b, ror=r, duration=pm.math.exp(log_duration))
    
    light_curve = 1e3 * starrystar.get_light_curve(orbit=orbit, t=t, r=r, texp=0.01)[:,0]
    
    # Simulate data
    y_true = pmx.eval_in_model(light_curve)
    random = np.random.default_rng(42)
    y = y_true + yerr * random.normal(size=len(t))
    plt.plot(t, y)
    plt.plot(t, y_true)

    # Likelihood
    pm.Normal("obs", mu=mean+light_curve, sigma=pm.math.sqrt(yerr**2 + pm.math.exp(log_jit)), observed=y)
```

```python
with model_dfm:
    trace_dfm = pmx.sample(tune=5000, draws=1000, chains=2, return_inferencedata=True)
```

```python
corner.corner(trace_dfm);
```

```python
time = np.copy(t)
flux = np.copy(y)
flux_err = np.copy(yerr)

plt.plot(time, flux)
```

# Define new basis

### $\gamma_1 = \frac{1}{3}b + \frac{2}{3}r$
### $\gamma_2 = \frac{2}{3}b - \frac{1}{3}r$

```python
r = trace_rand.posterior.r.values.reshape(-1)
b = trace_rand.posterior.b.values.reshape(-1)

y1 = (1/3)*b + (2/3)*r
y2 = (2/3)*b - (1/3)*r

corner.corner(np.vstack((y1, y2)).T, labels=["y1", "y2"]);
```

### Setting $r_{max}=1$ restricts ($\gamma_1, \gamma_2$) to a quadrangle with vertices at (0,0), (2/3, -1/3), (1/3, 2/3), and (4/3,1)

```python
# we'll define the y1 distribution using pm.Interpolated
x = np.linspace(0,1,N, endpoint=False)
xx = np.linspace(0,1,2*N, endpoint=False)
x_ = np.linspace(0,1,N, endpoint=True)
xx_ = np.linspace(0,1,2*N, endpoint=True)

Y1_x = np.hstack([x/3, (x+1)/3, (xx_+1)*2/3])
Y1_pdf = np.hstack([5/2*(x/3), 
                    5/6*(x/3) + 5/2/3, 
                   -5/3*(2*xx_/3) + 5/2/3 + 5/6/3])

Y2_x = np.hstack([(x-1)/3, xx*2/3, (x_+2)/3])
Y2_pdf = np.hstack([x, 1+xx*1e-8, 1-x_])


fig, ax = plt.subplots(1,2, figsize=(8,3))
ax[0].plot(Y1_x,Y1_pdf)
ax[0].set_title(r"$\gamma_1$")
ax[1].plot(Y2_x,Y2_pdf)
ax[1].set_title(r"$\gamma_2$")
plt.show()
```

```python
with pm.Model():
    y1 = pm.Interpolated("y1", Y1_x, Y1_pdf)
    
    lower = T.switch(T.lt(y1,2/3), -y1/2, 2*y1-5/3)
    upper = T.switch(T.lt(y1,1/3), 2*y1, y1/3+5/9)
    
    y2 = pm.Uniform("y2", lower=lower, upper=upper)
    
    b = pm.Deterministic("b", 3/5*y1 + 6/5*y2)
    r = pm.Deterministic("r", 6/5*y1 - 3/5*y2)
    
    pm.Potential("logr", -T.log(1+r))
    
    trace = pm.sample(draws=N, return_inferencedata=True)
```

```python
corner.corner(trace, var_names=["y1", "y2"]);
corner.corner(trace, var_names=["r", "b"]);
```

# Let's check the sampler performance

```python
with pm.Model() as model_gjg:

    # some model parameters
    mean = pm.Normal("mean")
    log_jit = pm.Normal("log_jit", mu=0.0, sigma=2.0)
    u = xo.QuadLimbDark("u")
    t0 = pm.Normal("t0", sigma=0.1)
    log_duration = pm.Uniform("log_duration", lower=np.log(0.01), upper=np.log(0.2), testval=np.log(0.1))
    
    # new r-b parameterization
    y1 = pm.Interpolated("y1", Y1_x, Y1_pdf)
    lower = T.switch(T.lt(y1,2/3), -y1/2, 2*y1-5/3)
    upper = T.switch(T.lt(y1,1/3), 2*y1, y1/3+5/9)
    y2 = pm.Uniform("y2", lower=lower, upper=upper)
    b = pm.Deterministic("b", 3/5*y1 + 6/5*y2)
    r = pm.Deterministic("r", 6/5*y1 - 3/5*y2)
    
    # this potential enforces a uniform prior on r
    pm.Potential("logr", -T.log(1+r))
    
    # build the lightcurve
    starrystar = xo.LimbDarkLightCurve(u)
    orbit = xo.orbits.KeplerianOrbit(period=10.0, t0=t0, b=b, ror=r, duration=pm.math.exp(log_duration))
    
    light_curve = 1e3 * starrystar.get_light_curve(orbit=orbit, t=time, r=r, texp=0.01)[:,0]
    
    # likelihood
    pm.Normal("obs", mu=mean+light_curve, sigma=pm.math.sqrt(flux_err**2 + pm.math.exp(log_jit)), observed=flux)
```

```python
with model_gjg:
    trace_gjg = pmx.sample(tune=5000, draws=1000, chains=2, return_inferencedata=True)
```

```python
corner.corner(trace_gjg);
```
