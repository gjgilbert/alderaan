```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
```

```python
def k0(r,b):
    return np.arccos((r**2 + b**2 - 1)/(2*r*b))

def k1(r,b):
    return np.arccos((1 - r**2 + b**2)/(2*b))

def z2(r,b):
    return np.sqrt((4*b**2 - (1 + b**2 - r**2)**2)/4)

def lam(r,b):
    return (1/pi)*(r**2*k0(r,b) + k1(r,b) -z2(r,b))

def ell(r,b):
    return -(r/2)*(b-r-1)
```

```python
eps = 1e-6

r = 0.1

b = np.linspace(1-r+eps, 1+r-eps)

plt.figure(figsize=(8,5))
plt.plot(b, lam(r,b), lw=2, label="Mandel & Agol 2002")
plt.plot(b, ell(r,b), lw=2, label="linear approximation")
plt.legend(fontsize=16)
plt.xlabel("b", fontsize=20)
plt.ylabel("overlap area", fontsize=20)
plt.show()
```
