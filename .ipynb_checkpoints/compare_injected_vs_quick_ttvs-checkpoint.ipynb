```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
import glob
```

```python
# manually set main directory and catalog files locations
PRIMARY_DIR  = "/Users/research/projects/alderaan/"

QUICK_TTV_DIR = PRIMARY_DIR + "QuickTTVs/"
SIM_TTV_DIR   = PRIMARY_DIR + "Simulations/TTVs/"
```

```python
sim_files = np.sort(glob.glob(SIM_TTV_DIR + "S*"))
quick_files = np.sort(glob.glob(QUICK_TTV_DIR + "S*/*quick*"))
```

```python
sim_planet_ids = []
quick_planet_ids = []

for i, sf in enumerate(sim_files):
    sim_planet_ids.append(sf[-22:-13])
    
for i, qf in enumerate(quick_files):
    quick_planet_ids.append(qf[-24:-15])
    
    
# only use the planets for which Quick TTVs have been computed
match = np.isin(sim_planet_ids, quick_planet_ids)

sim_files = sim_files[match]
sim_planet_ids = np.array(sim_planet_ids)[match]
```

```python
koi_ids = []

for i, pid in enumerate(sim_planet_ids):
    koi_ids.append("S" + pid[1:-3])
```

```python
for i, koi in enumerate(np.unique(koi_ids)):
    use = np.array(koi_ids) == koi
    NPL = np.sum(use)
    
    
    sfiles = sim_files[use]
    qfiles = quick_files[use]
    
    fig, ax = plt.subplots(NPL, 1, figsize=(12,6))
    
    for npl in range(NPL):
        sinds, stts = np.loadtxt(sfiles[npl]).swapaxes(0,1)
        qinds, qtts = np.loadtxt(qfiles[npl]).swapaxes(0,1)
        
        sephem = poly.polyval(sinds, poly.polyfit(sinds, stts, 1))
        qephem = poly.polyval(qinds, poly.polyfit(qinds, qtts, 1))
        
        somc = stts - sephem
        qomc = qtts - qephem
        
        ax[npl].plot(sephem, somc*24*60, "k")
        ax[npl].plot(qephem, qomc*24*60, "o", c="C{0}".format(npl))
        
    plt.suptitle(koi, fontsize=24)
    plt.show()
```
