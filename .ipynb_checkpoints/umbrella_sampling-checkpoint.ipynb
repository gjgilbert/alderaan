```python
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
import pymc3_ext as pmx
import arviz as az
import exoplanet as xo
import corner
import aesara_theano_fallback.tensor as T
import scipy.signal as sig
from scipy import stats

pi = np.pi
```

```python
N = 10000

rmin = 0.01
rmax = 0.95
```

# First let's set up the new basis for grazing transits

### $\lambda = r^2 + \beta r$
### $q = \beta/r$

In this basis, $\beta \equiv 1-b$.

This basis is derived from the geometric equations of Mandel & Agol 2002. When limb darkening is negligible, the geometric dependence of the fractional reduction in flux, $\lambda$, is nearly linear in $b$ over the range $1-r < b < 1+r$ for any constant $r < 1$. Generally, $\lambda$ is well constrained by the data. The second variable, $q$, was chosen to provide a convenient collective variable to use when applying the EMUS algorithm for umbrella sampling. The variable $q$ can be loosely interpreted as how strongly grazing a transit is.

### Draw samples from the standard basis

```python
with pm.Model():

    r = pm.Uniform("r", lower=rmin, upper=rmax)
    b = pm.Uniform("b", lower=1-r, upper=1+r)
        
    trace = pm.sample(draws=N*10, return_inferencedata=True)
```

```python
r = trace.posterior.r.values.reshape(-1)
b = trace.posterior.b.values.reshape(-1)

q = (1 - b)/r
lam = r**2 + (1-b)*r

corner.corner(np.vstack((r, b)).T, labels=["r", "b"]);
corner.corner(np.vstack((q, lam)).T, labels=["q", "lam"]);
```

```python
corner.corner(np.vstack((lam, q, r, 1-b)).T, labels=["lambda", "q", "r", "beta"]);
```

### Draw samples in the new basis and check that priors are correct

```python
with pm.Model():
    q = pm.Uniform("q", lower=-1, upper=1)
    lam = pm.Uniform("lam", lower=0, upper=(q+1)*rmax**2)
    
    r = pm.Deterministic("r", pm.math.sqrt(lam/(q+1)))
    beta = pm.Deterministic("beta", q*r)
    
    pm.Potential("log_r", -pm.math.log(r))
        
    trace = pm.sample(draws=N, return_inferencedata=True)
    
corner.corner(trace);
```

# Simulate a transit lightcurve and sample using the standard approach

```python
t = np.sort(np.random.default_rng(678).uniform(-0.1, 0.1, 500))
yerr = 10.0

with pm.Model() as model_standard:

    mean = pm.Normal("mean")
    log_jit = pm.Normal("log_jit", mu=0.0, sigma=5.0)
    
    u = xo.QuadLimbDark("u")
    log_r = pm.Uniform("log_r", lower=np.log(rmin), upper=np.log(rmax), testval=np.log(0.1))
    r = pm.math.exp(log_r)
    b = pm.Uniform("b", lower=0, upper=1+r, testval=0.86)
    
    t0 = pm.Normal("t0", sigma=0.1)
    log_dur = pm.Uniform("log_dur", lower=np.log(0.01), upper=np.log(0.2), testval=np.log(0.1))
    
    starrystar = xo.LimbDarkLightCurve(u)
    orbit = xo.orbits.KeplerianOrbit(period=10.0, t0=t0, b=b, ror=r, duration=pm.math.exp(log_dur))
    
    light_curve = 1e3 * starrystar.get_light_curve(orbit=orbit, t=t, r=r, texp=0.01)[:,0]
    
    # Simulate data
    y_true = pmx.eval_in_model(light_curve)
    random = np.random.default_rng(42)
    y = y_true + yerr * random.normal(size=len(t))
    plt.plot(t, y)
    plt.plot(t, y_true)

    # Likelihood
    pm.Normal("obs", mu=mean+light_curve, sigma=pm.math.sqrt(yerr**2 + pm.math.exp(log_jit)), observed=y)
```

```python
vars = pm.util.get_default_varnames(model_standard.unobserved_RVs, True)
truths = {
    var.name: value
    for var, value in zip(
        vars, model_standard.fastfn(vars)(model_standard.test_point)
    )
}
```

```python
with model_standard:
    trace_standard = pmx.sample(tune=2000, draws=1000, chains=2, return_inferencedata=True)
```

```python
corner.corner(trace_standard, truths=truths);
```

# Now try umbrella sampling

### Grazing umbrella (standard parameterization)

```python
with pm.Model() as model_us1:
    mean = pm.Normal("mean")
    log_jit = pm.Normal("log_jit", mu=0.0, sigma=5.0)    
    
    
    log_r = pm.Uniform("log_r", lower=np.log(rmin), upper=np.log(rmax), testval=np.log(0.1))
    r = pm.math.exp(log_r)
    b = pm.Uniform("b", lower=1-r, upper=1+r, testval=1.0)
    
    pm.Potential("psi", pm.Triangular.dist(lower=-1, c=0, upper=1).logp((1-b)/r))
        
    u = xo.QuadLimbDark("u")
    t0 = pm.Normal("t0", sigma=0.1)
    log_dur = pm.Uniform("log_dur", lower=np.log(0.01), upper=np.log(0.2), testval=np.log(0.1))
    
    starrystar = xo.LimbDarkLightCurve(u)
    orbit = xo.orbits.KeplerianOrbit(period=10.0, t0=t0, b=b, ror=r, duration=pm.math.exp(log_dur))
    
    light_curve = 1e3 * starrystar.get_light_curve(orbit=orbit, t=t, r=r, texp=0.01)[:,0]
    
    # Likelihood
    pm.Normal("obs", mu=mean+light_curve, sigma=pm.math.sqrt(yerr**2 + pm.math.exp(log_jit)), observed=y)
```

```python
with model_us1:
    trace_us1 = pmx.sample(tune=5000, draws=1000, chains=2, target_accept=0.95, return_inferencedata=True)
```

```python
corner.corner(trace_us1);
```

```python
az.summary(trace_us1)
```

### Grazing umbrella (new parameterization)

```python
with pm.Model() as model_us1:
    mean = pm.Normal("mean")
    log_jit = pm.Normal("log_jit", mu=0.0, sigma=5.0)    
    
    
    # drawing q from pm.Triangular auotmatically applies a tent bias psi
    #q = pm.Uniform("q", lower=2*(rmin/rmax)**2-1, upper=1)
    q = pm.Triangular("q", lower=2*(rmin/rmax)**2-1, c=0, upper=1)
    
    log_lam = pm.Uniform("log_lam", lower=T.log(2*rmin**2), upper=T.log((q+1)*rmax**2))
    lam = pm.Deterministic("lam", T.exp(log_lam))
        
    r = pm.math.sqrt(lam/(q+1))    
    b = pm.Deterministic("b", 1-q*r)
    log_r = pm.Deterministic("log_r", pm.math.log(r))
    
    # not exactly the jacobian because the boundary condiations on lambda already induce a conditional prior
    pm.Potential("pseudo_jacobian", -log_r)
    
    # we actually want the prior to be uniform in log_r, not log_lam
    pm.Potential("r_marginal", log_lam-log_r)

        
    
    u = xo.QuadLimbDark("u")
    t0 = pm.Normal("t0", sigma=0.1)
    log_dur = pm.Uniform("log_dur", lower=np.log(0.01), upper=np.log(0.2), testval=np.log(0.1))
    
    starrystar = xo.LimbDarkLightCurve(u)
    orbit = xo.orbits.KeplerianOrbit(period=10.0, t0=t0, b=b, ror=r, duration=pm.math.exp(log_dur))
    
    light_curve = 1e3 * starrystar.get_light_curve(orbit=orbit, t=t, r=r, texp=0.01)[:,0]
    
    # Likelihood
    pm.Normal("obs", mu=mean+light_curve, sigma=pm.math.sqrt(yerr**2 + pm.math.exp(log_jit)), observed=y)
```

```python
with model_us1:
    trace_us1 = pmx.sample(tune=5000, draws=1000, chains=2, target_accept=0.95, return_inferencedata=True)
```

```python
corner.corner(trace_us1);
```

```python
az.summary(trace_us1)
```

### Barely grazing umbrella

```python
with pm.Model() as model_usb:

    mean = pm.Normal("mean")
    log_jit = pm.Normal("log_jit", mu=0.0, sigma=5.0)
    
    
    
    
    log_r = pm.Uniform("log_r", lower=np.log(rmin), upper=np.log(rmax), testval=np.log(0.1))
    r = pm.math.exp(log_r)
    b_offset = pm.Bound(pm.Normal, lower=r-1, upper=2*r)("b_offset", mu=0, sd=0.1, testval=0)
    b = pm.Deterministic("b", 1-r+b_offset)

    pm.Potential("log_opr", -pm.math.log(1+r))
    
    
    
    
    u = xo.QuadLimbDark("u")
    t0 = pm.Normal("t0", sigma=0.1)
    log_dur = pm.Uniform("log_dur", lower=np.log(0.01), upper=np.log(0.2), testval=np.log(0.1))
    
    starrystar = xo.LimbDarkLightCurve(u)
    orbit = xo.orbits.KeplerianOrbit(period=10.0, t0=t0, b=b, ror=r, duration=pm.math.exp(log_dur))
    
    light_curve = 1e3 * starrystar.get_light_curve(orbit=orbit, t=t, r=r, texp=0.01)[:,0]
    
    # Likelihood
    pm.Normal("obs", mu=mean+light_curve, sigma=pm.math.sqrt(yerr**2 + pm.math.exp(log_jit)), observed=y)
```

```python
with model_usb:
    trace_usb = pmx.sample(tune=2000, draws=1000, chains=2, return_inferencedata=True)
```

```python
corner.corner(trace_usb);
```
