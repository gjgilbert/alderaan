```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
import scipy.stats as stats
from scipy.special import logsumexp
import scipy.linalg as linalg
from astropy.stats import freedman_bin_width

pi = np.pi
```

# First simulate a complicated potential

```python
Npdf = int(1e6)
#random = np.random.seed(222)
random = np.random.seed(234)

xpts = np.random.uniform(0,10, size=6)
ypts = np.random.uniform(3,6, size=6)

pfit = poly.polyfit(xpts,ypts,5)

x_pdf = np.linspace(0,10,Npdf)
y_pdf = poly.polyval(x_pdf, pfit)
#y_pdf = y_pdf - y_pdf.min() + 0.1
y_pdf = y_pdf/(np.sum(y_pdf)/Npdf*(x_pdf.max()-x_pdf.min()))

plt.plot(x_pdf, y_pdf);
```

# Define umbrella bias windows

```python
def psi_0(x):
    return 2*stats.triang(c=0.5, loc=-4, scale=8).pdf(x)

def psi_1(x):
    return stats.triang(c=0.5, loc=1, scale=8).pdf(x)

def psi_2(x):
    return 2*stats.triang(c=0.5, loc=6, scale=8).pdf(x)
```

```python
psi_fxns = [psi_0, psi_1, psi_2]

Nwin = len(psi_fxns)

for i in range(Nwin):
    plt.plot(x_pdf, psi_fxns[i](x_pdf))
```

# Draw (biased) samples from each window

```python
Nsamp = Npdf//10
samples = np.zeros((Nwin,Nsamp))

for i, psi in enumerate(psi_fxns):
    prob = psi(x_pdf)*y_pdf
    prob /= np.sum(prob)
    
    samples[i] = np.random.choice(x_pdf, p=prob, size=Nsamp)
    

bins = np.linspace(0,10,41)

fig, ax = plt.subplots(1,2, figsize=(12,4))

for i, samp in enumerate(samples):
    ax[0].hist(samp, bins=bins, density=True, histtype="step", lw=2)
    
ax[1].hist(np.hstack(samples), bins=bins, density=True, histtype="step", lw=2, color="grey")
ax[1].plot(x_pdf, y_pdf, "r")
```

# Now calculate weights

TODO: Incorporate autocorrelation time estimates

```python
def F_iter(psi_fxns, samples, z):
    Nwin = len(psi_fxns)
    F = np.zeros((Nwin,Nwin))
    
    for i in range(Nwin):

        denom = 0.
        for k in range(Nwin):
            denom += psi_fxns[k](samples[i])/z[k]

        for j in range(Nwin):
            num = psi_fxns[j](samples[i])/z[i]

            F[i,j] = np.mean(num/denom)
            
    return F
```

```python
def z_iter(F, tol=1.E-10, max_iter=100):   
    
    # stationary distribution is the last column of QR factorization
    M = np.eye(len(F))-F
    q,r = linalg.qr(M)
    z = q[:,-1] 
    z /= np.sum(z)
    
    # polish solution using power method.
    for itr in range(max_iter):
        znew = np.dot(z,F)
        tv = np.abs(znew[z > 0] - z[z > 0]) 
        tv = tv/z[z > 0]
        
        maxresid = np.max(tv) 
        if maxresid < tol:
            break
        else:
            z = znew
            
    # return normalized (by convention)
    return z/np.sum(z)
```

```python
z = np.zeros(Nwin)

for i, psi in enumerate(psi_fxns):
    z[i] = np.mean(psi(samples[i]))
    
z /= np.sum(z)
```

```python
nMBAR = 20
for n in range(nMBAR):
    F = F_iter(psi_fxns, samples, z)
    z = z_iter(F)
    
    # perform self-consistent polishing until convergence
    for n in range(nMBAR):
        z_old = np.copy(z)
        z_old[z_old < 1e-100] = 1e-100
        
        F = F_iter(psi_fxns, samples, z)
        z = z_iter(F)
        
        # check if we have converged
        if np.max(np.abs(z-z_old)/z_old) < 1e-8:
            break
```

# Resample and combine for the final histogram

The regions where windows don't overlap are undersampled in the combined histogram. This issue isn't present if the left edge of $\psi_{j+2}$ touches the right edge of $\psi_j$ so that there are no gaps where windows don't overlap.

```python
combo_samples = []
weights = []
for i in range(Nwin):
    combo_samples.append(samples[i])
    
    # compute weights as in eq 4 in Matthews et al. 
    wd = 0.
    for k in range(Nwin):
        wd += psi_fxns[k](samples[i])/z[k]

    weights.append(1./wd)
combo_samples = np.array(combo_samples).flatten()
weights = np.array(weights).flatten()

sub_samples = np.random.choice(combo_samples, p=weights/np.sum(weights), size=Nsamp)


# histogram using all weighted samples
fig, ax = plt.subplots(1,2, figsize=(13,4))

ax[0].hist(combo_samples, weights=weights, bins=bins, histtype="step", density=True, lw=2, color="grey", label='samples histogram')
ax[0].plot(x_pdf, y_pdf, label='target pdf', c='C0')
ax[0].legend(frameon=False, loc='best')
ax[0].set_title("all weighted samples")

# histogram using sub-samples
ax[1].hist(sub_samples, bins=bins, histtype="step", density=True, lw=2, color="grey", label='samples histogram')
ax[1].plot(x_pdf, y_pdf, label='target pdf', c='r')
ax[1].legend(frameon=False, loc='best')
ax[1].set_title("sub-sampled draws")

plt.show()
```
